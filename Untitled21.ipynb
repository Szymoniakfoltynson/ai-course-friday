```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_filter(grayscale_image, filter_kernel):

    #wymiar obrazu
    image_height, image_width = grayscale_image.shape

    #obraz wyjciowy
    filtered_image = np.zeros((image_height - 2, image_width - 2), dtype=np.float32)

    for y in range(1, image_height - 1):
        for x in range(1, image_width - 1):
          neighborhood = grayscale_image[y-1:y+2, x-1:x+2]
          convolution_value = np.sum(neighborhood * filter_kernel)
          filtered_image[y-1, x-1] = convolution_value

    filtered_image = np.abs(filtered_image)
    filtered_image = np.clip(filtered_image, 0, 255).astype(np.uint8)

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    axes[0].set_title("Oryginalny obraz w skali szaroci")
    axes[0].imshow(grayscale_image, cmap='gray')

    axes[1].set_title("Naoony filtr")
    axes[1].imshow(filtered_image, cmap='gray')

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import numpy as np
# import matplotlib.pyplot as plt # Already imported in apply_filter's defining cell

img = cv2.imread('contents_paragraph_center_horizontal_tmb_1606.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#wykrywanie krawdzi pionowych
apply_filter(gray, np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
]))
#wykrywanie krawdzi poziomych
apply_filter(gray, np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
]))
```

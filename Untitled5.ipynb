```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
tokenizer = AutoTokenizer.from_pretrained("flax-community/PapuGaPT2")
model = AutoModelForCausalLM.from_pretrained("flax-community/PapuGaPT2")
generator = pipeline('text-generation', model=model, tokenizer=tokenizer)
```

```python
text = generator("Szymon Foltyn to",
                 max_length=50,
                temperature= 0.9,
                top_p= 0.92,
                repetition_penalty =1.15,
                top_k = 60,
                do_sample=True)
print(text[0]['generated_text'])
```

```python
text = generator("Sisiek to odklejczyk",
                max_length = 20,
                num_beams = 6,
                do_sample = True,
                repetition_penalty = 1.2,
                early_stopping = True)
output = text[0]['generated_text']
last_dot = output.rfind(",")
if last_dot != -1:
  output = output[:last_dot+1]
print(output)
```

```python
import matplotlib.pyplot as plt

#Dane - skład powietrza (w procentach objętości)
skladniki = ['Azot (N2)', 'Tlen (O2)', 'Argon (Ar)', 'Dwutlenek węgla (CO2)', 'Inne gazy']
udzialy = [78.08, 20.95, 0.93, 0.04, 0.00]

plt.figure(figsize=(6,6))
plt.pie(udzialy, labels=skladniki, autopct='%1.2f%%', startangle=90)
plt.title('Skład powietrza atmosferycznego')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 12, 100)

v = 120 * (1 - np.exp(-t / 3))

plt.figure(figsize=(9,5))
plt.plot(t, v, 'r', linewidth=2, label='Prędkość samochodu')

plt.title('Rozpędzanie się samochodu', fontsize=14)
plt.xlabel('Czas [s]')
plt.ylabel('Prędkość [km/h]')
plt.grid(True)

predkosci_docelowe = [50, 100]
for target in predkosci_docelowe:

  idx = (np.abs(v - target)).argmin()
  plt.scatter(t[idx], v[idx], color='blue')
  plt.text(t[idx]+0.2, v[idx]-5, f'{target} km/h\npo {t[idx]:.1f} s', color='blue')

plt.legend()
plt.show()
```
